# Data Visualization and Descriptive Analysis

### Distribution plots ( satisfaction_level,last_evaluation, average_monthly_hours) 

```python

# Set up the matplotlib figure
f, axes = plt.subplots(ncols=3, figsize=(15, 5))

# Graph Employee Satisfaction
sns.distplot(df.satisfaction_level, kde=False, color="blue", ax=axes[0]).set_title('Employee Satisfaction Distribution')
axes[0].set_ylabel('Employee Count')

# Graph Employee Evaluation
sns.distplot(df.last_evaluation, kde=False, color="green", ax=axes[1]).set_title('Employee Evaluation Distribution')
axes[1].set_ylabel('Employee Count')

# Graph Employee Average Monthly Hours
sns.distplot(df.average_montly_hours, kde=False, color="red", ax=axes[2]).set_title('Employee Average Monthly Hours Distribution')
axes[2].set_ylabel('Employee Count')
```

### Distribution of employee count 

```python
features=['number_project','time_spend_company','Work_accident','left', 'promotion_last_5years','Departments ','salary']
fig=plt.subplots(figsize=(15,20))
for i, j in enumerate(features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace = 1.0)
    sns.countplot(x=j,data = df)
    plt.xticks(rotation=90)
    plt.title("No. of employee")
```

## Comparing the stats between employees who left and employees who stayed

```python
fig=plt.subplots(figsize=(15,20))
for i, j in enumerate(features):
    plt.subplot(4, 2, i+1)
    plt.subplots_adjust(hspace = 1.0)
    sns.countplot(x=j,data = df, hue='left')
    plt.xticks(rotation=90)
    plt.title("No. of employee")
```

# Cluster Analysis

```python
from sklearn.cluster import KMeans
left_emp = df[['satisfaction_level','last_evaluation']][df.left ==1]
kmeans = KMeans(n_clusters = 3,random_state = 0).fit(left_emp)
left_emp['label'] = kmeans.labels_
plt.scatter(left_emp['satisfaction_level'],left_emp['last_evaluation'],c = left_emp['label'], cmap = 'Accent')
plt.xlabel('Satisfaction Level')
plt.ylabel('Evaluation')
plt.title('Cluster analysis of employees who left')
plt.show()
```

High Satisfaction and High Evaluation(Shaded by green color in the graph), you can also call them Winners.

Low Satisfaction and High Evaluation(Shaded by blue color(Shaded by green color in the graph), you can also call them Frustrated.

Moderate Satisfaction and moderate Evaluation (Shaded by grey color in the graph), you can also call them 'Bad match'.
